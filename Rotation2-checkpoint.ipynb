## Optimal LMA for decidious plants:
For decidious plants, biomass assimilation rate is given by:

$ \large g_{de} = \frac{1}{LC}\left[A_{0}LL\left(1-\frac{LL k_{1}k_{2} v_{c.max25}}{2u LMA}\right) - C_{c}LMA \right] $ 

<br>
Assuming that LL is tightly constained by growing season such that $LL=fLC$, where $f$ is the fraction of the year suitable for growth, we obtain $g_{de}$ as a function of LMA. 
<br>
<br>
$\large g_{de}(LMA) = \left[A_{0}f\left(1-\frac{fLC k_{1}k_{2} v_{c.max25}}{2u LMA}\right) - C_{c}LMA \right] $
<br>
<br>
Optimizing $g_{de}$ with respect to $LMA$ we find that the maximal assimilation rate occurs when:
<br>
<br>
$ \large LMA = f\left(\frac{A_{0}LCk_{1}k_{2}v_{c.max25}}{2uC_{c}}\right)^{1/2} = LMA_{opt} $

## Scaling up to a distribution of leaves

Assuming LMA is distributed log-normally within a collection of leaves from either an individual or a plant community such that:
<br>
<br> 
$ \large f(LMA) = \frac{1}{LMA \sigma \sqrt{2 \pi}} exp\left[\frac{-\left(\ln{LMA} - \mu\right)^2}{2\sigma^{2}}\right]$
<br>
<br> 
Then, the average biomass assimilation rate is given by 
$ \large \left<g_{de}\right>_{LMA} = \int_{0}^{\infty} \left[A_{0}f\left(1-\frac{fLC k_{1}k_{2} v_{c.max25}}{2u x}\right) - C_{c}x \right] \frac{1}{x \sigma \sqrt{2 \pi}} exp\left[\frac{-\left(\ln{x} - \mu\right)^2}{2\sigma^{2}}\right] dx $
<br>
<br>
We define $\Gamma_{0} = A_{0}f $ and $\Gamma_{1} = \frac{A_{0}f^{2}LC k_{1}k_{2}v_{c.max25}}{2u}$ and rewrite the integral as
<br>
<br>
$ \large \left<g_{de}\right>_{LMA} = \int_{0}^{\infty} \left[\Gamma_{0}-\frac{\Gamma_{1}}{x} - C_{c}x \right] \frac{1}{x \sigma \sqrt{2 \pi}} exp\left[\frac{-\left(\ln{x} - \mu\right)^2}{2\sigma^{2}}\right] dx $
<br>
<br>
Solving this definite integral renders
<br>
<br>
$ \large \left<g_{de}\right>_{LMA} = \Gamma_{0} - \Gamma_{1}\exp(\frac{\sigma^{2}}{2} - \mu) - C_{c}\exp(\frac{\sigma^{2}}{2} + \mu)$
<br>
<br>
Optimizing for $\mu$ we find that $\left<g_{de}\right>_{LMA}$ is maximal when 
<br>
<br>
$ \large \ln(\mu) = \left(\frac{\Gamma_{1}}{C_{c}}\right)^{1/2} = LMA_{opt}$

## Constraining total leaf area using hydraulic traits and metabolic scaling
Total water flow rate is constrained by basal metabolism on the lower end, and available precipitation on the upper bound. Hence, basal flow rate is related to water use efficiency and metabolism through:
<br>
<br>
$ \large Q_{0} = \gamma B_{0} $
<br>
Where $\gamma$ is water use efficiency (mol H2O/mol C), and $B_{0}$ is basal metabolic rate (mol C/ time). Assuming that $B_{0}$  scales according to height by a some power law, we can rewrite $Q_{0}$ in terms of height.
<br>
$ \large Q_{0} = \beta_{hQ} h^{\eta_{hQ}}$
<br> 

### Available flow rate due to precipitation
To quantify the maximal available flow rate we take into account $\gamma$, incoming precipitation $P_{inc}$ (mol H20/$m^{2}$), and the disc area comprised by the roots $\pi r_{root}^{2}$. 
<br>
$ \large Q_{p} = \gamma \pi r_{root}^{2} P_{inc} $
<br>
### Total or realized flow rate
Total volume flow rate is given by evaporative molar flux $E$ (mol H2O/$m^{2}s$), the total area of stomatal openings $a_{f}$, and the molar mass and density of water.
<br>
$ \large Q_{e} = \frac{a_{f}E \mu_{W}}{\rho_{W}} $
<br>
$E$ can be found by solving the energy budget equation
<br>
$ \large R_{abs} - La_{g} - Ha_{j} - \lambda E a_{f} = 0 $
<br>

Following Kempes et al
<br>
$ \large E = \frac{f_{1}}{\lambda}\left[\frac{R_{abs} - f_{2}a_{f} + g_{2}a_{g}}{f_{1}a_{f} + g_{1}a_{g} + j_{1}a_{j}} \right] +\frac{f_{2}}{\lambda}$
<br>
Where $f_{1}$, $f_{2}$, $g_{1}$, $g_{2}$, and $j_{1}$ encode the effect of various environmental parameters and $R_{abs}$ is dependent on leaf dimensions, canopy geometry and incoming solar radiation. We note that $a_{g}$, $a_{f}$, and $a_{j}$ can all be written in terms of the total one-sided leaf area $a_{L}$, which can be related to LMA.
<br>
$ \large a_{L} = a_{l}n^{N} = \left<\frac{M_{l}}{LMA}\right>n^{N} $
<br>
We can assume that the average of leaf mass over LMA is well approximated by average leaf mass over average LMA.
<br>
$ \large a_{L} = \frac{n^{N}}{\left<LMA\right>}\frac{M_{L}}{n^{N}}$
<br> 
Hence, given a scaling relationship between photosynthetic biomass and total biomass, we can approximate one sided leaf area as:
<br>
$ \large a_{L} = \frac{\beta_{LM}M^{\eta_{LM}}}{\left<LMA\right>} $
<br> 
Substituting into $a_{g}$, $a_{f}$, and $a_{j}$ we get
<br>

$ \large a_{f} = \frac{2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}}{\left<LMA\right>}$
<br>
$ \large a_{g} = \frac{\beta_{LM}M^{\eta_{LM}}}{\left<LMA\right>}$
<br>
$ \large a_{j} = \frac{2\beta_{LM}M^{\eta_{LM}}}{\left<LMA\right>}$
<br>

$R_{abs}$ is defined as
<br>

$\large R_{abs} = \left[1 - \zeta_{c} - (1-\zeta_{s})\tau \right]P_{can}R_{inc} $
