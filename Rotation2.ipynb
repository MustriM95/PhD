## Optimal LMA for decidious plants:
For decidious plants, biomass assimilation rate is given by:

$ \large g_{de} = \frac{1}{LC}\left[A_{0}LL\left(1-\frac{LL k_{1}k_{2} v_{c.max25}}{2u LMA}\right) - C_{c}LMA \right] $ 

<br>
Assuming that LL is tightly constained by growing season such that $LL=fLC$, where $f$ is the fraction of the year suitable for growth, we obtain $g_{de}$ as a function of LMA. 
<br>
<br>
$\large g_{de}(LMA) = \left[A_{0}f\left(1-\frac{fLC k_{1}k_{2} v_{c.max25}}{2u LMA}\right) - C_{c}LMA \right] $
<br>
<br>
Optimizing $g_{de}$ with respect to $LMA$ we find that the maximal assimilation rate occurs when:
<br>
<br>
$ \large LMA = f\left(\frac{A_{0}LCk_{1}k_{2}v_{c.max25}}{2uC_{c}}\right)^{1/2} = LMA_{opt} $

## Scaling up to a distribution of leaves

Assuming LMA is distributed log-normally within a collection of leaves from either an individual or a plant community such that:
<br>
<br> 
$ \large f(LMA) = \frac{1}{LMA \sigma \sqrt{2 \pi}} exp\left[\frac{-\left(\ln{LMA} - \mu\right)^2}{2\sigma^{2}}\right]$
<br>
<br> 
Then, the average biomass assimilation rate is given by 
$ \large \left<g_{de}\right>_{LMA} = \int_{0}^{\infty} \left[A_{0}f\left(1-\frac{fLC k_{1}k_{2} v_{c.max25}}{2u x}\right) - C_{c}x \right] \frac{1}{x \sigma \sqrt{2 \pi}} exp\left[\frac{-\left(\ln{x} - \mu\right)^2}{2\sigma^{2}}\right] dx $
<br>
<br>
We define $\Gamma_{0} = A_{0}f $ and $\Gamma_{1} = \frac{A_{0}f^{2}LC k_{1}k_{2}v_{c.max25}}{2u}$ and rewrite the integral as
<br>
<br>
$ \large \left<g_{de}\right>_{LMA} = \int_{0}^{\infty} \left[\Gamma_{0}-\frac{\Gamma_{1}}{x} - C_{c}x \right] \frac{1}{x \sigma \sqrt{2 \pi}} exp\left[\frac{-\left(\ln{x} - \mu\right)^2}{2\sigma^{2}}\right] dx $
<br>
<br>
Solving this definite integral renders
<br>
<br>
$ \large \left<g_{de}\right>_{LMA} = \Gamma_{0} - \Gamma_{1}\exp(\frac{\sigma^{2}}{2} - \mu) - C_{c}\exp(\frac{\sigma^{2}}{2} + \mu)$
<br>
<br>
Optimizing for $\mu$ we find that $\left<g_{de}\right>_{LMA}$ is maximal when 
<br>

$ \large e^{\mu} = \left(\frac{\Gamma_{1}}{C_{c}}\right)^{1/2} = LMA_{opt}$
<br>

We note that $e^{\mu}$ in this case is the median of the distribution and not the mean.


## Constraining total leaf area using hydraulic traits and metabolic scaling
Total water flow rate is constrained by basal metabolism on the lower end, and available precipitation on the upper bound. Hence, basal flow rate is related to water use efficiency and metabolism through:
<br>
<br>
$ \large Q_{0} = \gamma B_{0} $
<br>
Where $\gamma$ is water use efficiency (mol H2O/mol C), and $B_{0}$ is basal metabolic rate (mol C/ time). Assuming that $B_{0}$  scales according to height by a some power law, we can rewrite $Q_{0}$ in terms of height.
<br>
$ \large Q_{0} = \beta_{hQ} h^{\eta_{hQ}}$
<br> 

### Available flow rate due to precipitation
To quantify the maximal available flow rate we take into account $\gamma$, incoming precipitation $P_{inc}$ (mol H20/$m^{2}$), and the disc area comprised by the roots $\pi r_{root}^{2}$. 
<br>
$ \large Q_{p} = \gamma \pi r_{root}^{2} P_{inc} $
<br>
### Total or realized flow rate
Total volume flow rate is given by evaporative molar flux $E$ (mol H2O/$m^{2}s$), the total area of stomatal openings $a_{f}$, and the molar mass and density of water.
<br>
$ \large Q_{e} = \frac{a_{f}E \mu_{W}}{\rho_{W}} $
<br>
$E$ can be found by solving the energy budget equation
<br>
$ \large R_{abs} - La_{g} - Ha_{j} - \lambda E a_{f} = 0 $
<br>

Following Kempes et al
<br>
$ \large E = \frac{f_{1}}{\lambda}\left[\frac{R_{abs} - f_{2}a_{f} + g_{2}a_{g}}{f_{1}a_{f} + g_{1}a_{g} + j_{1}a_{j}} \right] +\frac{f_{2}}{\lambda}$
<br>
Where $f_{1}$, $f_{2}$, $g_{1}$, $g_{2}$, and $j_{1}$ encode the effect of various environmental parameters and $R_{abs}$ is dependent on leaf dimensions, canopy geometry and incoming solar radiation. We note that $a_{g}$, $a_{f}$, and $a_{j}$ can all be written in terms of the total one-sided leaf area $a_{L}$, which can be related to LMA.
<br>
$ \large a_{L} = a_{l}n^{N} = \left<\frac{M_{l}}{LMA}\right>n^{N} $
<br>
We can assume that the average of leaf mass over LMA is well approximated by average leaf mass over the average reciprocal of LMA.
<br>
$ \large a_{L} = n^{N}\left<\frac{1}{LMA}\right>\frac{M_{L}}{n^{N}}$
<br> 
Hence, given a scaling relationship between photosynthetic biomass and total biomass, we can approximate one sided leaf area as:
<br>
$ \large a_{L} = \beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right> $
<br> 
Substituting into $a_{g}$, $a_{f}$, and $a_{j}$ we get
<br>

$ \large a_{f} = 2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>$
<br>
$ \large a_{g} = \beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>$
<br>
$ \large a_{j} = 2\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>$
<br>

$R_{abs}$ is defined as
<br>

$\large R_{abs} = \left[1 - \zeta_{c} - (1-\zeta_{s})\tau \right]P_{can}R_{inc} $
<br>

Where canpoy reflection $\zeta_{c}$, and transmission $\tau$ depend on leaf area index (LAI).
<br>

$\large LAI = \frac{a_{l}n^{N}}{\pi r_{can}^{2}} = \frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}} $
<br>

Substituting into $\tau$ and $\zeta_{c}$:
<br>

$ \large \tau = \exp\left[-\sqrt{\alpha}K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{\pi r_{can}^{2}}\right] $
<br>
And
<br>
$ \large \zeta_{c} = \frac{\zeta_{c}^{*} + f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]}{1 + \zeta_{c}^{*}f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]} $
<br>

Substituting into $R_{abs}$:

$ \large R_{abs} = \left[1 - \frac{\zeta_{c}^{*} + f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]}{1 + \zeta_{c}^{*}f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]} - (1 - \zeta_{s})\exp\left[-\sqrt{\alpha}K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{\pi r_{can}^{2}}\right] \right]P_{can}R_{inc}$

Finally, we get an expression for $Q_{e}$ as a function of $\left<LMA\right>$.
<br>

$ \large Q_{e} = \frac{\mu_{W}}{\rho_{W}}\left(2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>\right) \left\{\frac{f_{1}}{\lambda}\left[\frac{R_{abs}\left\{\left<LMA\right>\right\} - f_{2}\left(2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>\right) + g_{2}\left(\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right)}{f_{1}\left(2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>\right) + g_{1}\left(\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right) + j_{1}\left(2\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right)} \right] +\frac{f_{2}}{\lambda}\right\} $
<br>

We can further note that $f_{1}$, $f_{2}$, and $j_{1}$ depend on leaf area through $g_{\upsilon}$ and $g_{Ha}$.

$  g_{\upsilon} = \frac{1}{g_{\upsilon L}^{-1} + g_{\upsilon a}^{-1}}$,  $ g_{Ha} = \frac{0.664 \hat{\rho}_{a}D_{H}R_{e_{a}}^{1/2}P_{r_{a}}^{1/3}}{d}$, and $  g_{\upsilon a} = \frac{0.664 \hat{\rho}_{a}D_{\nu}R_{e_{a}}^{1/2}S_{c_{a}}^{1/3}}{d}$
<br>
Where $ d=1.62(\frac{a_{l}}{\pi})^{1/2} =1.62\left(\frac{\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right> }{\pi}\right)^{1/2}$

And $R_{e_{a}} = \frac{ud}{\nu_{a}} = \hat{R}_{e_{a}}d$, with $u$ being wind speed and $\nu_{a}$ the kinematic viscosity of air.
<br>
To simplify the algebra we make the LMA dependence of $d$ explicit by defining $d = \hat{d}\left<LMA^{-1}\right>^{1/2}$. Substituting into $f_{1}$, $f_{2}$, and $j_{1}$ we get:
<br>

$\large j_{1} = c_{p}g_{Ha} = c_{p}\frac{0.664 \hat{\rho}_{a}D_{H}\hat{R}_{e_{a}}^{1/2}P_{r_{a}}^{1/3}}{\hat{d}^{1/2}\left<LMA^{-1}\right>^{1/4}} = \frac{c_{p}\hat{g}_{Ha}}{\left<LMA^{-1}\right>^{1/4}} = \frac{\hat{j}_{1}}{\left<LMA^{-1}\right>^{1/4}}$
<br>

$ \large f_{1} = \lambda \frac{d e_{s}}{dT_{A}} \frac{0.664\hat{\rho}_{a}D{\nu}\hat{R}_{e_{a}}^{1/2}S_{c_{a}}^{1/3}a_{s}\delta_{s}}{\hat{d}^{1/2}\left<LMA^{-1}\right>^{1/4}a_{s}\delta_{s} + 0.664R_{e_{a}}^{1/2}S_{c_{a}}^{1/3}z_{s}}$
<br>

$ \large f_{2} = \lambda \frac{D_{\nu}}{p_{a}} \frac{0.664\hat{\rho}_{a}D{\nu}\hat{R}_{e_{a}}^{1/2}S_{c_{a}}^{1/3}a_{s}\delta_{s}}{\hat{d}^{1/2}\left<LMA^{-1}\right>^{1/4}a_{s}\delta_{s} + 0.664R_{e_{a}}^{1/2}S_{c_{a}}^{1/3}z_{s}}$





## Incorporating hydraulic constraints into assimilation rate
We can describe the total biomass assimilation rate by multiplying average area-specific assimilation $\left<g_{de}\right>_{LMA}$ by total one-sided leaf area $a_{L}$.
<br>

$\large G_{de} = a_{L}\left<g_{de}\right>_{LMA} = \left(\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right)\left[\Gamma_{0} - \Gamma_{1}\exp(\frac{\sigma^{2}}{2} - \mu) - C_{c}\exp(\frac{\sigma^{2}}{2} + \mu)\right] $
<br>

More generally, for an arbitrary LMA distribution.
<br>

$\large G_{de} = \left(\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right)\left[\Gamma_{0} - \Gamma_{1}E\left<\frac{1}{LMA}\right> - C_{c}\left<LMA\right>\right] $
<br>

We can incorporate the effect of water limitation by multiplying gross biomass production (disregarding the cost of maintenance) by a saturating function of the ratio of evaporative flow rate to available flow rate $Q_{e}/Q_{p}$.

$\large G_{de} = \left(\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right)\left[\left(\frac{Q_{p}}{\kappa Q_{e} + Q_{p}}\right)^{\upsilon}\left(\Gamma_{0} - \Gamma_{1}\left<\frac{1}{LMA}\right>\right) - C_{c}\left<LMA\right>\right] $
<br>

Where $\kappa$ is a half saturation constant and $\upsilon$ is a shape parameter that determines how sensitive growth is to water limitation. For $\upsilon << 0$, water scarcity is only an issue as $Q_{e} \approx Q_{p}$, if $\upsilon \approx 1$ there is an inverse decrease, and for $\upsilon > 1$ water limitation has a drastic impact. We can generally set $\upsilon = 1$, and determine $\kappa$ by noting that the height/biomass at which $G_{de} = 0$ is the theoretical maximum height. Additionally, we can incorporate the effects of canopy geometry into the assimilation rate by rescaling total leaf area by the proportion of intercepted PPFD, which can be calculated using the canopy absorption coefficient for photosynthetically active radiation $I_{c}$, and the canopy projected area $P_{can}$.

$\large I_{c} = (1-\zeta_{c})\left\{1-\exp \left[-K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]\right\}$

Substituting into our expression, we get:

$\large G_{de} = \beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\left(\frac{Q_{p}}{\kappa Q_{e} + Q_{p}}\right)\left(\Gamma_{0} - \Gamma_{1}\left<LMA^{-1}\right>\right)\left(I_{c}P_{can}\right) - \beta_{0}M^{\eta_{0}}  $

```python
import math
import numpy as np
from matplotlib import pyplot as plt
from pylab import meshgrid
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from scipy import optimize
```

```python
def R_abs(eSLA, T_A = 20, p_a = 101.3, M=10, p_inc=0.940, uw=4, psi=0.1, T_D=0, R_inc=1380, f_g=0.75):
    
    #constants
    mu_w = 18.01528 # g/mol
    rho_w = 1 # g/cm^{3}
    gamma = 1/3 # WUE
    kappa = 1.0 #eWUE
    D_H = 18.46*10**(-6) # Thermal heat diffusivity of air (m^{2}/s)
    D_nu = 24.9*10**(-6) # Thermal heat diffusivity of water in air (m^{2}/s)
    nu_a = 1.52*10**(-5) # kinematic viscosity of air (m^2/s) at 20C
    S_ca = nu_a/D_nu # Dimensionless Schmidt number 
    P_ra = nu_a/D_H # Dimensionless Prandtl number for air
    c_p = 29.10 # Specific heat of air (isobaric) (J/mol K)
    sig = 5.67*10**(-8) # Stefan-Boltzmann Constant (W/m^2K^4)
    b_1 = 0.611 # KPa
    b_2 = 17.502 # Dimensionless
    b_3 = 240.97 # deg C
    lamb = 40660 # latent heat of water evaporation (J/mol)
    
    # Scaling
    bet_lm = 0.16 # scaling coefficient for leaf mass from total biomass
    eta_lm = 0.779 #scaling exponent for leaf mass from total biomass
    bet_hm = 2.58
    eta_hm = 0.264
    bet_3 = 0.423
    bet_6 = 0.9
    bet_5 = 0.3524
    eta_5 = 1.14
    bet_mr = 0.421 # micro mol C/s*kg^eta_mr scaling coefficient for total plant respiration
    eta_mr = 0.78 # scaling exponent for total plant respiration
    # 
    delta_s = 220 #stomatal density stomata/mm^2
    a_s = 235.1*10**(-6) # stomatal area mm^2
    z_s = 10*10**(-6) # stomatal depth m
    eps_l = 0.95# leaf emissivity ()
    b_1 = 0.611 # KPa
    b_2 = 17.502 # Dimensionless
    b_3 = 240.97 # deg C
    lamb = 40660 # latent heat of water evaporation (J/mol)
    # Environmental dependencies of leaf conductance
    rho_a = (44.6*p_a*273.15)/(101.3*T_A) # molar density of air (mol/m^3)
    d = 1.62*np.sqrt((bet_lm*M**(eta_lm)*eSLA)/math.pi)
    R_ea = uw*d/nu_a
    e_a = b_1*np.exp(b_2*T_A/(b_3+T_A))
    de_a = (b_1*b_2*b_3/(b_3+T_A)**(2))*math.exp(b_2*T_A/(b_3+T_A))
    D_v = b_1*np.exp(b_2*T_A/(b_3+T_A)) - b_1*np.exp(b_2*T_D/(b_3+T_D))
    
    # Conductances
    g_ua = (0.664*rho_a*D_v*(R_ea**(1/2))*(S_ca**(1/3)))/d
    g_us = rho_a*D_nu/z_s
    g_ul = g_us*a_s*delta_s
    g_ups = 1/((1/g_ul) + (1/g_ua))
    g_Ha = (0.664*rho_a*D_H*(R_ea**(1/2))*(P_ra**(1/3)))/d
    g_r = 4*eps_l*sig*T_A**(3)/c_p # Radiative conductance
    
    # Flux coefficients
    g_1 = eps_l*sig*T_A**(4)
    g_2 = g_r*c_p
    j_1 = c_p*g_Ha
    f_1 = lamb*g_ups*de_a/p_a
    f_2 = lamb*g_ups*D_v/p_a
    
    # Flux coefficients
    g_1 = eps_l*sig*T_A**(4)
    g_2 = g_r*c_p
    j_1 = c_p*g_Ha
    f_1 = lamb*g_ups*de_a/p_a
    f_2 = lamb*g_ups*D_v/p_a
    a_f = 2*bet_lm*M**(eta_lm)*delta_s*a_s*eSLA
    a_g = bet_lm*M**(eta_lm)*eSLA
    a_j = 2*bet_lm*M**(eta_lm)*eSLA

    # Canopy geomery 
    h = bet_hm*M**(eta_hm) # Tree height in m
    r_can = bet_5*h**(eta_5) # Canopy radial extent in m
    h_can = bet_6*h**(eta_5) # Canopy height in m
    zeta_dc = 0.06 # Deep canopy refelction coefficient
    zeta_s = 0.30 # Soil reflection coefficient
    alph = 0.5 # leaf absorptivity

    # Canopy projections
    theta = np.arctan(-h_can/(2*r_can*np.tan(psi)))
    P_can = math.pi*r_can*(r_can*np.cos(theta) + 0.5*h_can*np.sin(theta)*np.tan(-psi))
    S_arc = np.arcsin(2*math.sqrt((0.5*h_can)**(2) - r_can**(2))/h_can)
    S_can = 2*np.pi*(r_can**(2) + (r_can**(2)*(h_can*0.5)**(2))*S_arc/(np.sqrt((0.5*h_can)**(2) - r_can**(2))))
    K = 2*P_can/S_can
    
    LAI = bet_lm*M**(eta_lm)*eSLA/(math.pi*r_can**(2)) # Leaf area index
    f = (zeta_dc-zeta_s)/(zeta_dc*zeta_s - 1) 
    zeta_c = (zeta_dc + f*np.exp(-2*K*LAI))/(1 + zeta_dc*f*np.exp(-2*K*LAI)) # Canopy reflectance
    tau = np.exp(-np.sqrt(alph)*K*LAI) # Canopy transmission ceofficient
    alph_can = 1 - zeta_c - (1-zeta_s)*tau # Total solar radiaition absorption coefficient for canopy
    I_c = (1-zeta_c)*(1-np.exp(-K*LAI)) # PAR absorption coefficient for canopy
    R_abs = ((1 - zeta_c) - (1-zeta_s)*tau)*P_can*R_inc
    
    return I_c
```

```python
def G_de(eSLA, T_A = 20, p_a = 101.3, M=10, p_inc=0.940, uw=4, psi=0.5*np.pi, T_D=0, R_inc=1380, f_g=0.75, LC=365, kappa=1, kap=1):
    
    #constants
    JtoMol = 4.6
    fPAR = 0.5
    mu_w = 0.01801528 # kg/mol
    rho_w = 998 # kg/m^{3}
    gamma = 1/3 # WUE
    D_H = 18.46*10**(-6) # Thermal heat diffusivity of air (m^{2}/s)
    D_nu = 24.9*10**(-6) # Thermal heat diffusivity of water in air (m^{2}/s)
    nu_a = 1.52*10**(-5) # kinematic viscosity of air (m^2/s) at 20C
    S_ca = nu_a/D_nu # Dimensionless Schmidt number 
    P_ra = nu_a/D_H # Dimensionless Prandtl number for air
    c_p = 29.10 # Specific heat of air (isobaric) (J/mol K)
    sig = 5.67*10**(-8) # Stefan-Boltzmann Constant (W/m^2K^4)
    b_1 = 0.611 # KPa
    b_2 = 17.502 # Dimensionless
    b_3 = 240.97 # deg C
    lamb = 40660 # latent heat of water evaporation (J/mol)
    
    # Scaling
    n = 2
    bet_lm = 0.16 # scaling coefficient for leaf mass from total biomass
    eta_lm = 0.779 #scaling exponent for leaf mass from total biomass
    bet_hm = 2.58
    eta_hm = 0.264
    bet_3 = 0.423
    bet_6 = 0.9
    bet_5 = 0.3524
    eta_5 = 1.14
    bet_mr = 0.421 # micro mol C/s*kg^eta_mr scaling coefficient for total plant respiration
    eta_mr = 0.78 # scaling exponent for total plant respiration
    # 
    delta_s = 220 #stomatal density stomata/mm^2
    a_s = 235.1*10**(-6) # stomatal area mm^2
    z_s = 10*10**(-6) # stomatal depth m
    eps_l = 0.95# leaf emissivity ()
    b_1 = 0.611 # KPa
    b_2 = 17.502 # Dimensionless
    b_3 = 240.97 # deg C
    lamb = 40660 # latent heat of water evaporation (J/mol)
    
    a_L = bet_lm*M**(eta_lm)*eSLA
    nN = n**(np.log(M))
    
    # Environmental dependencies of leaf conductance
    rho_a = (44.6*p_a*273.15)/(101.3*T_A) # molar density of air (mol/m^3)
    d = 1.62*np.sqrt(a_L/(nN*np.pi))
    R_ea = uw*d/nu_a
    e_a = b_1*np.exp(b_2*T_A/(b_3+T_A))
    de_a = (b_1*b_2*b_3/(b_3+T_A)**(2))*np.exp(b_2*T_A/(b_3+T_A))
    D_v = b_1*np.exp(b_2*T_A/(b_3+T_A)) - b_1*np.exp(b_2*T_D/(b_3+T_D))
    
    # Conductances
    g_ua = (0.664*rho_a*D_v*(R_ea**(1/2))*(S_ca**(1/3)))/d
    g_us = rho_a*D_nu/z_s
    g_ul = g_us*a_s*delta_s
    g_ups = 1/((1/g_ul) + (1/g_ua))
    g_Ha = (0.664*rho_a*D_H*(R_ea**(1/2))*(P_ra**(1/3)))/d
    g_r = 4*eps_l*sig*T_A**(3)/c_p # Radiative conductance
    
    # Flux coefficients
    g_1 = eps_l*sig*T_A**(4)
    g_2 = g_r*c_p
    j_1 = c_p*g_Ha
    f_1 = lamb*g_ups*de_a/p_a
    f_2 = lamb*g_ups*D_v/p_a
    
    # Flux coefficients
    g_1 = eps_l*sig*T_A**(4)
    g_2 = g_r*c_p
    j_1 = c_p*g_Ha
    f_1 = lamb*g_ups*de_a/p_a
    f_2 = lamb*g_ups*D_v/p_a
    a_f = 2*bet_lm*M**(eta_lm)*delta_s*a_s*eSLA
    a_g = bet_lm*M**(eta_lm)*eSLA
    a_j = 2*bet_lm*M**(eta_lm)*eSLA

    # Canopy geomery 
    h = bet_hm*M**(eta_hm) # Tree height in m
    r_can = bet_5*h**(eta_5) # Canopy radial extent in m
    h_can = bet_6*h**(eta_5) # Canopy height in m
    zeta_dc = 0.06 # Deep canopy refelction coefficient
    zeta_s = 0.30 # Soil reflection coefficient
    alph = 0.5 # leaf absorptivity

    # Canopy projections
    theta = np.arctan(-h_can/(2*r_can*np.tan(psi)))
    P_can = math.pi*r_can*(r_can*np.cos(theta) + 0.5*h_can*np.sin(theta)*np.tan(-psi))
    S_arc = np.arcsin(2*np.sqrt((0.5*h_can)**(2) - r_can**(2))/h_can)
    S_can = 2*np.pi*(r_can**(2) + (r_can**(2)*(h_can*0.5)**(2))*S_arc/(np.sqrt((0.5*h_can)**(2) - r_can**(2))))
    K = 2*P_can/S_can
    
    LAI = (bet_lm*M**(eta_lm))*eSLA/(math.pi*r_can**(2)) # Leaf area index
    f = (zeta_dc-zeta_s)/(zeta_dc*zeta_s - 1) 
    zeta_c = (zeta_dc + f*np.exp(-2*K*LAI))/(1 + zeta_dc*f*np.exp(-2*K*LAI)) # Canopy reflectance
    tau = np.exp(-np.sqrt(alph)*K*LAI) # Canopy transmission ceofficient
    alph_can = 1 - zeta_c - (1-zeta_s)*tau # Total solar radiaition absorption coefficient for canopy
    I_c = (1-zeta_c)*(1-np.exp(-K*LAI)) # PAR absorption coefficient for canopy
    R_abs = ((1 - zeta_c) - (1-zeta_s)*tau)*P_can*R_inc
    
    # Hydraulic traits
    r_roo = bet_3**(1/4)*h
    Q_p = gamma*(math.pi*r_roo**2)*p_inc
    Q_e = (mu_w/rho_w)*a_f*((f_1/lamb)*((R_abs - a_f*f_2 - a_g*g_2)/(f_1*a_f + g_1*a_g + j_1*a_j)) + f_2/lamb)
    
    # Leaf photosynthesis
    u = 768 # Dimensionless +- 71
    k1 = 0.030 # Scaling factor (Kg of biomass/molC)
    k2 = 0.0864 # Scaling factor (s*molC/day*micromolC) = (molC/day)/(micromolC/s)
    Kc_25 = 0.03997 # kPa
    E_aKc = 79430 # J/mol
    Ko_25 = 27.480 # kPa
    E_aKo = 36380 # J/mol
    po2 = 0.2095*p_a
    c_a = 0.0004*p_a
    T_K = T_A + 273.15
    k_b = 8.314 #J/K*mol
    a_l = 0.8 #leaf absorptance
    b_l = 0.010875# fraction of a_l that reaches photsystem II
    T_crt = 647.096 # Critical temperature of water at sea level K
    H0 = 1.677
    H1 = 2.204
    H2 = 0.636
    H3 = -0.241
    
    def arrh(T_C, E_a):
        T_K = T_C + 273.15
        k_b = 8.314 #J/K*mol
        return np.exp(-E_a/(k_b*T_K))/np.exp(-E_a/(k_b*273.15))

    def hT(T_C):
        T_K = T_C + 273.15
        k_b = 8.314 #J/K*mol
        a_s = 668.39 #K*mol
        b_s = 1.07 # J/mol K^2
        dS = a_s - b_s*T_K
        H_d = 200000 # J/mol
        H_v = 71513 # J/mol
        farrh = np.exp(-H_v/(k_b*T_K))/np.exp(-H_v/(k_b*273.15))
        mod = (1+np.exp((273.15*dS - H_d)/(k_b*273.15)))/(1+np.exp((T_K*dS - H_d)/(k_b*T_K)))
        return farrh*mod


    Kc = Kc_25*arrh(T_C=T_A, E_a=E_aKc)
    Ko = Ko_25*arrh(T_C=T_A, E_a=E_aKo)
    K_s = Kc*(1+(po2/Ko)) # Effective Michaelis-Menten Coefficient
    beta_carb = 146 # ratio of unit costs for the maintenance of carboxylation and water transport capacities at 25C

    eta_s = 100*np.sqrt(T_K/T_crt)/(H0 + H1*(T_crt/T_K) + H2*(T_crt/T_K)**2 + H3*(T_crt/T_K)**3) # viscosity of water depends on temp
    phi_0 = a_l*b_l*(0.352 + 0.022*T_A - 0.00034*T_A**2)/4# intrinsic quantum efficiency of photosynthesis (dep T)
    I_abs = R_inc*fPAR*JtoMol # Leaf absorbed PPFD
    gam_s = po2*np.exp(6.779 - 37830/(T_K*k_b)) # Photorespiratory compensation point (depT)

    zet = np.sqrt(beta_carb*(gam_s+K_s)/(1.6*eta_s))
    c_i = (zet*c_a +gam_s*np.sqrt(D_v))/(zet + np.sqrt(D_v)) # Leaf internal partial pressure of CO2
    m = (c_i - gam_s)/(c_i + 2*gam_s)
    m_c = (c_i - gam_s)/(c_i + K_s)
    Vcmax_gt = phi_0*I_abs*m/m_c
    Vcmax_25 = Vcmax_gt/hT(T_C=T_A)
    A_0 = k1*k2*Vcmax_gt*m_c
    
    B_0 = bet_mr*M**(eta_mr)*k1*k2
    gamma0 = A_0*f_g
    gamma1 = (A_0*f_g**(2))*LC*k1*k2*Vcmax_25/(2*u)
    
    Q_e = (Q_e*3600*8*365*f_g)
    
    #W = 1 + (0.8*(Q_r - 1.3)**2)*(((np.exp(Q_r))/((np.exp(Q_r) + kappa)))**2 -1)
    
    G_de = P_can*((Q_p/(Q_e*kap + Q_p))**kappa)*(gamma0 - gamma1*eSLA)*(I_c) - B_0
    
    #G_de = P_can*W*(gamma0 - gamma1*eSLA)*(I_c) - B_0
    
    return G_de
```
